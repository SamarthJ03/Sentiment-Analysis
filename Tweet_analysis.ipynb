```python
x_train = df['content']
x_test = df_t['content']
y_train = df['sentiment']
y_test = df_t['sentiment']
```

```python
df['sentiment'].value_counts()
```

**Preprocessing the data**

Tokenizing the data

```python
from keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token='<nothing>')
```

```python
x_train
```

```python
s = []

for i in range(x_train.size):
    s.append(x_train[i])
```

```python
s = [str(item) for item in s]

tokenizer.fit_on_texts(s)
```

```python
x_train.astype(str)
```

```python
len(tokenizer.word_index)
```

```python
x_train_strings = [str(item) for item in x_train]

# Convert text data to sequences
sequences = tokenizer.texts_to_sequences(x_train_strings)
```

```python
x_test_strings = [str(item) for item in x_test]

# Convert text data to sequences
xtest = tokenizer.texts_to_sequences(x_test_strings)
```

```python
from keras.utils import pad_sequences
```

```python
X_train = pad_sequences(sequences,padding='post',maxlen=50)
X_test = pad_sequences(xtest,padding='post',maxlen=50)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the categorical data
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)
```

```python
from keras.utils import to_categorical
y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)
```

**Model Building and Training**

```python
from keras import Sequential
from keras.layers import *
```

```python
from keras.models import Sequential
from keras.layers import Bidirectional, LSTM, Dense
import numpy as np


model = Sequential()
model.add(Embedding(33785,2,input_length=50))
model.add(Bidirectional(LSTM(32,return_sequences=True)))
model.add(Bidirectional(LSTM(32, return_sequences=True)))
model.add(Bidirectional(LSTM(32, return_sequences=True)))
model.add(Bidirectional(LSTM(32, return_sequences=False)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(4, activation='softmax'))


dummy_input = np.random.rand(1, 50, 1)
model(dummy_input)

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


model.summary()
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train_encoded,epochs=10,validation_data=(X_test,y_test_encoded))
```
